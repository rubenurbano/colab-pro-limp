```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Seno", color="blue")
plt.title("Gráfico de la función seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Combinador Y en Python (recursión sin nombres)
### 🧠 Explicación del Combinador Y en Python
La celda de código define un **Combinador Y**, que permite la **recursión en funciones anónimas** (`lambda`) sin nombrarlas explícitamente.

- `fact(n)` → calcula el factorial de `n`
- `fib(n)` → calcula el número `n` de Fibonacci

**Ejemplo de salida:**

Y = (lambda f: (lambda x: f(lambda *a: x(x)(*a)))
                 (lambda x: f(lambda *a: x(x)(*a))))

fact = Y(lambda rec: lambda n: 1 if n < 2 else n * rec(n - 1))
fib  = Y(lambda rec: lambda n: n if n < 2 else rec(n - 1) + rec(n - 2))

print("fact(6) =", fact(6))   # 720
print("fib(10) =", fib(10))   # 55
```
